```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
numeric_df = df.select_dtypes(include=np.number).copy()
numeric_df
```

```python
numeric_df.dtypes
```

```python
numeric_df.mean()
```

```python
# Replace missing values with the mean of each column
numeric_df = numeric_df.apply(lambda col: col.fillna(col.mean()))
```

```python
# Ensure we select only numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Fill missing values in numeric columns with the mean
df[numeric_cols] = df[numeric_cols].astype(float).apply(lambda col: col.fillna(col.mean()))
```

```python
print(df.isnull().sum())  # Check missing values in all columns
```
